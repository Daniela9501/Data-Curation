1. Identify and read the data table containing the number of people attending world cups from 1930 to 2018. Attendance for how many world cups is there in the table? Assume each row contains information about a unique world cup. 

```python
# Iterate through the tables to find the one with attendance information
attendance_table = None
for table in tables:
    # Check if the table contains columns related to World Cup attendance
    if "Year" in table.columns and "Hosts" in table.columns:
        attendance_table = table
        break

if attendance_table is not None:
    # Count the number of rows in the attendance table
    num_world_cups = len(attendance_table)
    print(f"Attendance for {num_world_cups} World Cups is included in the table.")
else:
    print("No table containing World Cup attendance information was found.")
```
